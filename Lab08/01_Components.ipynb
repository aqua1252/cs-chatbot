## Components of an Agentic AI System

A modern agentic system typically includes:

1. LLM Core

The reasoning engine (GPT-5, GPT-4o-mini, etc.) that interprets instructions.

2. Memory

Agents often need one or more memory types:
- Short-term scratchpad
- Long-term vector database

3. Tools / Actions

Agents perform actions such as:
- Running Python code
- Reading/writing files
- Calling APIs
- Querying databases
- Using browser automation
- Sending messages or emails
- Running shell commands (where safe)

4. Planning & Control Logic

Most frameworks provide one or more:
- ReAct loops (reason, act, observe)
- Tree-of-thought planners
- Workflow graphs
- Multi-agent coordination
