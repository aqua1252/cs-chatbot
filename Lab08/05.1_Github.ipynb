## ✅ 1. How to Create a Pull Request (PR)

A pull request is a way to propose changes to someone’s repository.

Steps:

1. Fork the repository

    - Go to the repo page.
    - Click Fork (top-right corner).
    - This creates your own copy of the repo.

2. Clone your fork
    ```
    git clone https://github.com/your-username/repo-name.git
    ```

3. Make changes
    - Edit files in your fork (either locally or in GitHub’s web editor).

4. Commit the changes
    - Locally:
        ```
        git add .
        git commit -m "Your message here"
        git push
        ```

5. Create the Pull Request
    - Go to your fork on GitHub.
    - Click Contribute > Open pull request.
    - Write a description and submit it.
---
## ✅ 2. How to Open an Issue

Issues are used to report problems, ask for features, or discuss ideas.

Steps:
1. Go to the repository on GitHub.
2. Click the Issues tab.
3. Click New issue.
4. Fill in the title and description.
5. Submit.

Note: Some repos disable issues. If you don’t see the tab, the owner turned it off.

---
## ✅ 3. How to Comment

You can comment on:
1. Issues
2. Pull requests
3. Code in PRs
4. Commits

Steps:
1. Open the issue or pull request.
2. Scroll to the bottom where it says Leave a comment.
3. Type your comment.
4. Click Comment.
