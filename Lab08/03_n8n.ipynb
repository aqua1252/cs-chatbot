## No-Code / Low-Code Agentic Systems

### n8n

n8n is a workflow automation tool that:
- Allows drag-and-drop visual flows
- Integrates with API nodes, AI nodes, webhooks
- Lets you embed LLMs and create simple agents
- No-code “agent-like” business automation

Similar No-Code Platforms:
- Zapier
- Make.com
- Dify
- FlowiseAI
